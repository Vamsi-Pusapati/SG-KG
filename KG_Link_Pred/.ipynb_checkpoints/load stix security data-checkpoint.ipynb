```python
k = None
for i in objs:
    if(i.get('type')=="attack-pattern"):
        
        k=i
        break
k.keys()
```

In the attack patterns we will be storing the:
id,
type,
external_references,
name,
description,
x_mitre_detection


```python
from neo4j import GraphDatabase
import json
import os

#uri = "bolt://localhost:7687"
#username = "neo4j"
#password = "neo4jkngbq"
uri="bolt://localhost:7687"
username = "neo4j"
password = "nimda123"
database = "neo4j"


def add_attack(tx, file_path):
    with open(filepath) as jsonfile:
        data = json.load(jsonfile)
        for obj in data['objects']:
            if(obj.get('type')=="attack-pattern"):
                
                query = """
                    CREATE (s:Attack {
                        id: $obj_id,
                        type: $obj_type,
                        external_references: $external_references,
                        name: $name,
                        description: $description,
                        x_mitre_detection: $x_mitre_detection
                    })
                    """
                
                tx.run(query, 
                   obj_id=obj.get('id'),
                   obj_type=obj.get('type'),
                   external_references=str(obj.get('external_references')),
                   name=obj.get('name'),
                   description=obj.get('description'),
                   x_mitre_detection=obj.get('x_mitre_detection'))
                
                print("create Security node of type - " + obj.get('type') + " with name - " + obj.get('name'))
                
    
    

try:
    file_path = './attack-stix-data/enterprise-attack/enterprise-attack.json'
    with GraphDatabase.driver(uri, auth=(username, password), database=database) as driver:
        with driver.session() as session:
            session.write_transaction(add_attack, file_path)

except Exception as e:
    print(f"Error: {e}")
```

```python
from neo4j import GraphDatabase
import json


uri = "bolt://localhost:7687"
username = "neo4j"
password = "nimda123"
database = "neo4j"

# Batch size
BATCH_SIZE = 500

def add_relations_in_batches(tx, relationships):
    for relationship in relationships:
        relation = relationship.get('relationship_type').replace('-', '_')
        query = f"""
        MATCH (a {{id: $source_ref}})
        MATCH (b {{id: $target_ref}})
        MERGE (a)-[r:{relation}]->(b)
        """
        tx.run(query,
               source_ref=relationship.get('source_ref'),
               target_ref=relationship.get('target_ref'))

def process_file_in_batches(file_path):
    with open(file_path) as jsonfile:
        data = json.load(jsonfile)
        relationships = [obj for obj in data['objects'] if obj.get('type') == "relationship" and obj.get('relationship_type') is not None]
        
        total_relationships = len(relationships)
        print(f"Total relationships to process: {total_relationships}")
        
        for i in range(0, total_relationships, BATCH_SIZE):
            batch = relationships[i:i + BATCH_SIZE]
            with GraphDatabase.driver(uri, auth=(username, password)) as driver:
                with driver.session() as session:
                    session.write_transaction(add_relations_in_batches, batch)
            print(f"Processed batch {i // BATCH_SIZE + 1} of {total_relationships // BATCH_SIZE + 1}")

try:
    file_path = './attack-stix-data/enterprise-attack/enterprise-attack.json'
    process_file_in_batches(file_path)

except Exception as e:
    print(f"Error: {e}")
```

# Delete all records

```python
from neo4j import GraphDatabase
import json
import os

#uri = "bolt://localhost:7687"
#username = "neo4j"
#password = "neo4jkngbq"
uri="bolt://localhost:7687"
username = "neo4j"
password = "nimda123"
database = "neo4j"


def clear_database(tx):
    tx.run("MATCH (n) DETACH DELETE n")

try:
    with GraphDatabase.driver(uri, auth=(username, password), database=database) as driver:
        with driver.session() as session:
            session.write_transaction(clear_database)

except Exception as e:
    print(f"Error: {e}")
```
